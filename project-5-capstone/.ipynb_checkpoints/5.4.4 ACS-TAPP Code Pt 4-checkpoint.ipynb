```python
list(Dem_08.columns.values)
```

```python
socialselectDem = pd.concat((Dem_16.iloc[:, :34], Dem_12.iloc[:, :34], Dem_08.iloc[:, :34]), axis=0)
socialselectDem
```

```python
Dem = pd.concat((Dem_16['Democrat Votes'], Dem_12['Democrat Votes'], Dem_08['Democrat Votes']), axis=0)
Dem
```

```python
socialselectDem = pd.concat((socialselectDem, Dem), axis=1)
socialselectDem
```

```python
list(socialselectDem.columns)
```

```python
# target = 2016 Democrat Votes by Select Social

yDemVotes = socialselectDem['Democrat Votes']
```

```python
data = pd.DataFrame(socialselectDem.iloc[:, :-1])
XDem_train, XDem_test, yDem_train, yDem_test = train_test_split(data, yDemVotes, test_size=0.5)
```

```python
print(XDem_train.shape)
XDem_train
```

```python
print(XDem_test.shape)
XDem_test
```

```python
yDem_train
```

```python
yDem_test
```

```python
slrDem = LinearRegression()
slrDem.fit(XDem_train, yDem_train)
```

```python
print(slrDem.score(XDem_test, yDem_test))
print(slrDem.score(XDem_train, yDem_train))
```

```python
slrDem.score(XDem_test, yDem_test)
```

```python
predsDem = slrDem.predict(XDem_test)
```

```python
plt.scatter(yDem_test, predsDem)
```

```python
scores = cross_val_score(slrDem, data, yDemVotes, cv=4)
print('Cross-validated scores =', scores)

predictions = cross_val_predict(slrDem, data, yDemVotes, cv=4)
plt.scatter(yDemVotes, predictions)

r2 = metrics.r2_score(yDemVotes, predictions)
print('Cross-Predicted R2 =', r2)
```
