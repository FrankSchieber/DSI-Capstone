```python
list(Rep_08.columns.values)
```

```python
socialselectRep = pd.concat((Rep_16.iloc[:, :34], Rep_12.iloc[:, :34], Rep_08.iloc[:, :34]), axis=0)
socialselectRep
```

```python
Rep = pd.concat((Rep_16['Republican Votes'], Rep_12['Republican Votes'], Rep_08['Republican Votes']), axis=0)
Rep
```

```python
socialselectRep = pd.concat((socialselectRep, Rep), axis=1)
socialselectRep
```

```python
list(socialselectRep.columns)
```

```python
# target = 2016 Republican Votes by Select Social

yRepVotes = socialselectRep['Republican Votes']
```

```python
data = pd.DataFrame(socialselectRep.iloc[:, :-1])
XRep_train, XRep_test, yRep_train, yRep_test = train_test_split(data, yRepVotes, test_size=0.5)
```

```python
print(XRep_train.shape)
XRep_train
```

```python
print(XRep_test.shape)
XRep_test
```

```python
yRep_train
```

```python
yRep_test
```

```python
slrRep = LinearRegression()
slrRep.fit(XRep_train, yRep_train)
```

```python
print(slrRep.score(XRep_test, yRep_test))
print(slrRep.score(XRep_train, yRep_train))
```

```python
slrRep.score(XRep_test, yRep_test)
```

```python
predsRep = slrRep.predict(XRep_test)
```

```python
plt.scatter(yRep_test, predsRep)
```

```python
scores = cross_val_score(slrRep, data, yRepVotes, cv=4)
print('Cross-validated scores =', scores)

predictions = cross_val_predict(slrRep, data, yRepVotes, cv=4)
plt.scatter(yRepVotes, predictions)

r2 = metrics.r2_score(yRepVotes, predictions)
print('Cross-Predicted R2 =', r2)
```
